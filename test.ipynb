```python
from citylearn.citylearn import CityLearnEnv
from agents.EVs.maddpg import MADDPGOptimizedRBC as RLAgent
from citylearn.reward_function import RewardFunction
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.signal
from typing import Any, List, Mapping, Union
```

```python
class CustomReward(RewardFunction):
    """Calculates custom multi-agent reward that considers energy sharing.

    Rewards buildings that share energy and penalizes high grid import/emissions.

    Parameters
    ----------
    env_metadata: Mapping[str, Any]
        General static information about the environment.
    """

    def __init__(self, env_metadata: Mapping[str, Any]):
        super().__init__(env_metadata)

    def calculate(self, observations: List[List[float]]) -> List[float]:
        """Calculates reward, considering energy sharing and emissions.

        Parameters
        ----------
        observations: List[List[float]]
            Nested list where each sublist contains numerical observations for a building.

        Returns
        -------
        reward: List[float]
            Reward values for each building.
        """

        # Extract the correct indices for each variable
        # Ensure these indices match the order of features in your observations list
        net_electricity_consumption_emission_idx = self.env.observation_order.index("net_electricity_consumption_emission")
        max_shared_energy_idx = self.env.observation_order.index("max_shared_energy")
        net_electricity_consumption_idx = self.env.observation_order.index("net_electricity_consumption")

        # Extract values from the observations list correctly
        net_electricity_consumption_emission = [o[net_electricity_consumption_emission_idx] for o in observations]
        shared_energy = [o[max_shared_energy_idx] for o in observations]
        grid_import = [o[net_electricity_consumption_idx] for o in observations]

        if self.central_agent:
            total_emissions = sum(net_electricity_consumption_emission)
            total_shared_energy = sum(shared_energy)
            reward = [-total_emissions + total_shared_energy]  # Reward sharing, penalize emissions
        else:
            reward = [
                shared - grid * 0.5 - emission
                for shared, grid, emission in zip(shared_energy, grid_import, net_electricity_consumption_emission)
            ]

        return reward
```

```python
# SIMULATION SETUP, TRAINING, & TESTING
env = CityLearnEnv(
    train,
    central_agent=False,
    reward_function=lambda env: CustomReward(env)
)


model = RLAgent(
    env,
    critic_units=[512, 256, 128],
    actor_units=[256, 128, 64],
    lr_actor=0.0006343946342268605,
    lr_critic=0.0009067117952187151,
    gamma=0.9773507798877807,
    sigma=0.2264587893937525,
    steps_between_training_updates=20,
    target_update_interval=100
)
```

```python
observations = env.reset()
```

```python
for time_step in range(env.time_steps - 1):
        actions = model.predict(observations)
        observations, _, _, _ = env.step(actions)
```
